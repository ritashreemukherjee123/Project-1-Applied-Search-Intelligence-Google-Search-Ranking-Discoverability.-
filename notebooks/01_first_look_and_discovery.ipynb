## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

```python
# Your discovery here

visible_df = df[df["impressions_90d"] > 0]
corr_all = df["search_volume"].corr(df["impressions_90d"])
corr_visible = visible_df["search_volume"].corr(visible_df["impressions_90d"])

print(f"Correlation (all pages): {corr_all:.3f}   (n={len(df):,})")
print(f"Correlation (impressions_90d > 0): {corr_visible:.3f}   (n={len(visible_df):,})")
print(f"\nDropped {len(df) - len(visible_df):,} pages with zero impressions.")
```

**Discovery E: In Discovery B, which content_type has the worst CTR at the same position tier?**


Redoing Discovery B, just verified  within same position tier (all tiers) mostly content type "comparison article" has the worst CTR.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos_type = visible.groupby(["position_tier", "content_type"])["ctr"].mean().unstack().round(4)
print(ctr_by_pos_type.to_string())
print("\n" + ctr_by_pos_type.idxmin(axis=1).to_string())
ctr_by_pos_type.plot(kind="bar", title="Mean CTR by position tier and content type", ylabel="CTR");
```

**Discovery F: Find another belief to test: does content_age_days relate to trend_direction? Does avg_position relate to ctr?**

Decline rate tends to decrease with content age, suggesting newer content might be more volatile.


A strong negative correlation implies CTR decreases as average position worsens.

```python
import pandas as pd, numpy as np

print("Belief 1: Content Age vs. Trend Direction")
base_age_bins = [0, 90, 180, 365, 730]
age_labels = ["0-3mo", "3-6mo", "6-12mo", "1-2yr", "2yr+"]

max_content_age_plus_one = df["content_age_days"].max() + 1
age_bins = list(base_age_bins)

if max_content_age_plus_one > age_bins[-1]:
    age_bins.append(max_content_age_plus_one)
else:
    age_bins.append(age_bins[-1] + 1)

df["age_bucket"] = pd.cut(df["content_age_days"], bins=age_bins, labels=age_labels, right=False)

decline_rate_by_age = (
    df.groupby("age_bucket")["trend_direction"]
      .apply(lambda x: (x == "down").mean())
)
overall_rate = (df["trend_direction"] == "down").mean()

print(f"Decline rate by age bucket:\n{decline_rate_by_age.round(3).to_string()}\n"
      f"Overall decline rate: {overall_rate:.3f}")

print("Belief 2: Average Position vs. CTR")
visible_for_ctr = df[df["impressions_90d"] >= 100]

corr_pos_ctr = visible_for_ctr["avg_position"].corr(visible_for_ctr["ctr"])
print(f"Correlation (avg_position vs. CTR for pages with >= 100 impressions): {corr_pos_ctr:.3f}")
```

The avg_position is a continuous numerical metric, and position_tier is a categorical grouping derived from it. A scatter plot is suitable for observing the relationship between two continuous variables, such as avg_position and ctr.


**The second plot, a scatter plot of 'Average Position vs. CTR,' implies a negative relationship between the two variables. As the avg_position increases (meaning the content ranks lower on the search results page), the 'Click-Through Rate (CTR)' tends to decrease.**

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot for Belief 1: Content Age vs. Trend Direction
decline_rate_by_age.plot(kind='bar', ax=axes[0], title='Decline Rate by Content Age Bucket', ylabel='Decline Rate')
axes[0].axhline(y=overall_rate, color='r', linestyle='--', label=f'Overall Decline Rate ({overall_rate:.3f})')
axes[0].set_xlabel('Content Age Bucket')
axes[0].set_ylabel('Mean Decline Rate')
axes[0].tick_params(axis='x', rotation=45)
axes[0].legend()

# Plot for Belief 2: Average Position vs. CTR
visible_for_ctr.plot(kind='scatter', x='avg_position', y='ctr', ax=axes[1], title='Average Position vs. CTR', alpha=0.6)
sample_size = min(len(visible_for_ctr), 5000)
visible_for_ctr.sample(sample_size).plot(kind='scatter', x='avg_position', y='ctr', ax=axes[1], title='Average Position vs. CTR', alpha=0.6)
axes[1].set_xlabel('Average Position')
axes[1].set_ylabel('Click-Through Rate (CTR)')

plt.tight_layout()
plt.show()
```
